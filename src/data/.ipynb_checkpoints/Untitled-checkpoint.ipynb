```python
import pandas as pd

# =========================
# 1. 读取数据
# =========================
# 如果你是从 js 导出的（export const places = [...]）
# 需要先去掉前缀，假设你已经处理成 json 或 dict
# 如果你已经是 df，这一步可以跳过


df = pd.read_json('places_full.json')



# =========================
# 2. 检查 card_name 问题
# =========================
mask_osm = df['card_name'].str.startswith('osm', na=False)
mask_null = df['card_name'].isna()

mask_problem = mask_osm | mask_null

print("====== 基本统计 ======")
print("总数据量:", len(df))
print("osm 开头:", mask_osm.sum())
print("null:", mask_null.sum())
print("总异常:", mask_problem.sum())


# =========================
# 3. 查看异常数据
# =========================
problem_df = df[mask_problem].copy()

print("\n====== 异常样本 ======")
display(problem_df[['place_id', 'card_name', 'name']].head(10))


# =========================
# 4. 自动修复逻辑
# =========================
# 优先级：
# 1. 用 name（通常更干净）
# 2. 如果 name 也没有 → 用 place_id
# 3. 去掉 osm_xxx 这种垃圾值

def fix_card_name(row):
    card = row['card_name']
    name = row.get('name')

    if pd.isna(card) or str(card).startswith('osm'):
        if pd.notna(name):
            return name
        else:
            return row['place_id']
    return card

df['card_name_fixed'] = df.apply(fix_card_name, axis=1)


# =========================
# 5. 替换原字段（可选）
# =========================
df['card_name'] = df['card_name_fixed']
df.drop(columns=['card_name_fixed'], inplace=True)


# =========================
# 6. 再检查一次
# =========================
mask_osm_after = df['card_name'].str.startswith('osm', na=False)
mask_null_after = df['card_name'].isna()

print("\n====== 清洗后 ======")
print("osm 开头:", mask_osm_after.sum())
print("null:", mask_null_after.sum())

```
